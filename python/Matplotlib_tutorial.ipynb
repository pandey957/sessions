```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = [1,2,3]
Y = [4,8,6]
X2 = [1,2,3]
Y2 = [10,14,12]
plt.plot(X,Y,label='first line')
plt.plot(X,Y2,label='second line')
plt.xlabel('Value of X Increasing')
plt.ylabel('Value of Y Increasing')
plt.legend()
plt.title('X \nY')
```

```python
x = [1,3,5,7,9,11]
y = [2,3,5,6,1,2]
x2 = [0,2,4,6,8,10]
y2 = [3,2,9,1,1,3]
```

```python
plt.bar(x,y,label='bar1',color='red')
plt.bar(x2,y2,label='bar2',color='blue')
plt.legend()
```

```python
sleeping = [2,3,3,1,5,8]
playing = [2,13,13,21,15,18]
eating = [11,13,23,11,5,9]
office = [12,1,2,3,1,8]
plt.plot([],[],label='sleeping',color = 'm',linewidth= 5)
plt.plot([],[],label='playing',color = 'b',linewidth= 5)
plt.plot([],[],label='eating',color = 'k',linewidth= 5)
plt.plot([],[],label='office',color = 'g',linewidth= 5)
plt.stackplot(x,sleeping,playing,eating,office,colors = ['m','b','k','r','g'])
plt.legend()
```
